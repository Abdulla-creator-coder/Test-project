```python
cols = ['business_partner_name','vehicle_no','vehicle_model','invoice_line_text']

for col in cols:
    freq = df[col].value_counts()
    df[col + '_encoded'] = df[col].map(freq)
```

```python
df.info()
```

```python
monthly_demand = df.groupby(
    [df['invoice_date'].dt.to_period('M'),
     'invoice_line_text_encoded']
).size().reset_index(name='demand')
```

```python
monthly_demand['invoice_date'] = monthly_demand['invoice_date'].dt.to_timestamp()
```

```python
spare_mapping = df[['invoice_line_text',
                    'invoice_line_text_encoded']].drop_duplicates()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

forecast_results = []

for spare in monthly_demand['invoice_line_text_encoded'].unique():

    spare_df = monthly_demand[
        monthly_demand['invoice_line_text_encoded'] == spare
    ].copy()

    spare_df.set_index('invoice_date', inplace=True)

    spare_df = spare_df.asfreq('MS')   #monthly frequency
    spare_df['demand'] = spare_df['demand'].fillna(0)

    try:
        model = SARIMAX(spare_df['demand'],
                        order=(1,1,1),
                        seasonal_order=(1,1,1,12))

        results = model.fit(disp=False)

        forecast = results.forecast(steps=6)  # next 6 months

        forecast_df = pd.DataFrame({
            'invoice_line_text_encoded': spare,
            'forecast_month': forecast.index,
            'forecast_demand': forecast.values
        })

        forecast_results.append(forecast_df)

    except:
        continue
```

```python
final_forecast = pd.concat(forecast_results)
```

```python
final_forecast = final_forecast.merge(
    spare_mapping,
    on='invoice_line_text_encoded',
    how='left'
)
```

```python
final_output = final_forecast[[
    'forecast_month',
    'invoice_line_text',
    'forecast_demand'
]]

print(final_output.head())
```
